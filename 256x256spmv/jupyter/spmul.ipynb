## 准备工作

生成稀疏矩阵

```python
import numpy as np
import random
sparse=0.35
size=256 #生成方阵的大小：size*size
matrix=np.arange(size*size).reshape(size,size)
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        x=random.randint(0,100)
        if x > (100*sparse):
            matrix[i][j]=0
        else:
            matrix[i][j]=x+1
print("Generate random matirx")
print(matrix)
```

普通矩阵转CSR矩阵函数

```python
# matrix to csr format
def matrix2csr(matrix):
    row_num=0
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] != 0:
                row_num+=1
#     values=np.arange(row_num).reshape(1,row_num)
#     colIndex=np.arange(row_num).reshape(1,row_num)
#     rowPtr=np.arange(len(matrix)+1)
    values=[1]*row_num
    colIndex=[1]*row_num
    rowPtr=[1]*(len(matrix)+1)
    k=0
    l=0
    tmp_size=list(range(len(matrix)))
    for i in tmp_size:
        for j in tmp_size:
            if j==0:
                rowPtr[l]=k
                l+=1
            if matrix[i][j]!=0:
                values[k]=matrix[i][j]
                colIndex[k]=j
                k+=1
    rowPtr[l]=row_num
    return values,colIndex,rowPtr
```

将普通稀疏矩阵转为CSR型稀疏矩阵

```python
values=[]
rowPtr=[]
colIndex=[]
values,colIndex,rowPtr=matrix2csr(matrix)
```

```python
# 数据备份
mat=matrix
val=values
col=colIndex
row=rowPtr
number=vec
```

```python
# 保存测试结果的变量
original_res=[1]*(len(number))
csr_res=[1]*(len(number))
```

```python
def matrix_mul_vector(matrix,x,y):
    n=list(range(len(matrix)))
    for i in n:
        y0=0.0
        for j in n:
            y0+=matrix[i][j]*x[j]
        y[i]=y0
```

```python
def spmv(rowPtr,colIndex,values,x,y):
    num_rows=list(range(len(x)))
    for i in num_rows:
        y0=0.0
        k=rowPtr[i]
        while k<rowPtr[i+1]:
            y0+=values[k]*x[colIndex[k]]
            k+=1
        y[i]=y0
```

```python
# 使用普通矩阵乘法
matrix_mul_vector(mat,number,original_res)
print(original_res)
```

```python
# 使用CSR乘法计算
spmv(row,col,val,number,csr_res)
print(csr_res)
```

## 加载Overlay模块
使用overlay模块进行ARM CPU与FPGA PL部分的交互

```python
from pynq import Overlay
overlay=Overlay("spmv.bit")
spmul=overlay.spmv_0
```

## 分配内存供IP使用
`使用pynq.allocate`函数分配内存空间

```python
from pynq import allocate
values_len=len(values)
rowPtr_len=len(rowPtr)
colIndex_len=len(colIndex)
vec_len=len(vec)

# 开辟存储空间
values_buffer=allocate(shape=(values_len,),dtype=np.float32)
vec_buffer=allocate(shape=(vec_len,),dtype=np.float32)
output_buffer=allocate(shape=(vec_len,),dtype=np.float32)
col_buffer=allocate(shape=(colIndex_len,),dtype=np.int32)
row_buffer=allocate(shape=(rowPtr_len,),dtype=np.int32)
```

将数据复制到刚开辟的内存中

```python
np.copyto(values_buffer,np.float32(values))
np.copyto(vec_buffer,np.float32(vec))
np.copyto(col_buffer,np.int32(colIndex))
np.copyto(row_buffer,np.int32(rowPtr))
```

```python
print(values_buffer)
print(vec_buffer)
print(col_buffer)
print(row_buffer)
print(output_buffer)
```

```python
print(values_buffer.physical_address)
print(output_buffer.physical_address)
```

使用IP的`write`方法，写入到IP对应的位置上

```python
spmul.s_axi_control.write(0x10,row_buffer.physical_address)
spmul.s_axi_control.write(0x1c,col_buffer.physical_address)
spmul.s_axi_control.write(0x28,values_buffer.physical_address)
spmul.s_axi_control.write(0x40,vec_buffer.physical_address)
spmul.s_axi_control.write(0x34,output_buffer.physical_address)
```

## 启动IP
启动位于0x00的控制信号

```python
import time
spmul.s_axi_CTRL.write(0x00,0x01)
while True:
    reg=spmul.s_axi_CTRL.read(0x00)
    if reg != 1:
        break
print("运行结束")
```

## 比较计算结果

```python
if len(output_buffer)!=len(csr_res):
    print("长度不等，计算结果有误")
length_list=list(range(len(csr_res)))
equal=True
for i in length_list:
    if output_buffer[i]!=csr_res[i]:
        equal=False

if equal:
    print("计算结果正确")
else:
    print("计算结果错误")
```
